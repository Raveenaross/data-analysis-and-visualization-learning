# Modifying DataFrames

## Renaming Columns and Index Labels

### `rename` function

This method is used to **rename labels** (columns or index) in a DataFrame.

#### Parameters
* `mapper` : Dict-like or function transformations to apply to that axis values.
* `index` : Dict-like or function to rename row labels.
* `columns`: Dict-like or function to rename column labels.
* `axis`: Specify whether to rename rows (`axis=0`) or columns (`axis=1`).
* `copy`: If False, tries to avoid copying data when possible.
* `inplace`: If True, modifies the DataFrame in place without returning a new one.
* `level`: Used when the index/columns are MultiIndex.
* `errors`: `ignore` skips labels not found.`raise` throws an error if a label doesn’t exist.

```python
import pandas as pd
houses = pd.read_csv("./materials/data/kc_house_data.csv")
titanic = pd.read_csv("./materials/data/titanic.csv")
netflix = pd.read_csv("./materials/data/netflix_titles.csv", sep="|", index_col=0)
btc = pd.read_csv("./materials/data/coin_Bitcoin.csv")
countries = pd.read_csv("./materials/data/world-happiness-report-2021.csv")
```

```python
countries.head(2)
```

```python
countries.set_index("Country name", inplace=True)
```

```python
mapper = {"Regional indicator":"regional_indicator", "Ladder score":"ladder_score"}
countries.rename(columns=mapper,inplace=True)
```

```python
countries.head(2)
```

```python
mapper= {"Finland":"finland", "Denmark":"denmark"}
countries.rename(index=mapper,inplace=True)
```

## `replace` method

Values of the Series/DataFrame are replaced with other values dynamically.

### Parameters
* `to_replace`:Defines what to replace. Can be:
   * Scalar (int, float, str)
   * List of values
   * Dict (mapping old → new values, optionally per column)
   * Regex (pattern matching)
   * None (special cases with regex or nested dicts)
* `value`:What to replace with. Can be scalar, list, dict, str, or regex.
* `inplace`: If `True`, modifies the DataFrame directly
* `regex`: Whether to interpret to_replace/value as regex.

```python
titanic.replace({"sex": {"female": "F", "male":"M"}}, inplace=True)
```

```python
titanic
```

```python
## titanic["sex"].replace(["Female", "Male"], ["F","M"])
```

```python
titanic["age"].replace("?", "IDK").value_counts()
```

```python
titanic["age"].replace(["?"], [None]).value_counts(dropna=False)
```

```python
titanic["age"] = titanic["age"].replace("?", None)
```

```python
titanic
```

## Making Updates with `.loc()` and Boolean Masks

```python
many_bedrooms = houses["bedrooms"] >= 10
```

```python
houses.loc[many_bedrooms, ['bedrooms', 'bathrooms']]
```

```python
houses.loc[many_bedrooms, ['bedrooms']] = "TOO MANY!"
```

```python
houses[many_bedrooms]
```

```python
high_quality = houses["grade"] > 12
great_views = houses["view"] == 4
houses[great_views & high_quality]
```

```python
houses["luxury_level"] = "normal"
```

```python
houses.head(5)
```

```python
houses.loc[high_quality & great_views, ['luxury_level']] = 'ultra'
```

```python
houses[great_views & high_quality]
```

```python
houses["luxury_level"].value_counts()
```
