# Filtering Data

## Filtering DataFrames with a Boolean Series

`df[bools]`: When you pass a list of Booleans to a pandas DataFrame, it acts as a row filter.

Each Boolean corresponds to a row in df.

`True` → keep the row.

`False` → drop the row.

```python
import pandas as pd
houses = pd.read_csv("./materials/data/kc_house_data.csv")
titanic = pd.read_csv("./materials/data/titanic.csv")
netflix = pd.read_csv("./materials/data/netflix_titles.csv", sep="|", index_col=0)
```

```python
titanic.sex == 'female'
```

```python
titanic[titanic.sex == 'female']
```

```python
titanic[titanic.survived == 0]
```

```python
df = titanic.head()
```

```python
bools = [True, False, True, True, True]
df[bools]
```

## Filtering with Comparison operators

Filtering the dataframes using comparison operators such as `==`, `!=`, `<`, `>`, `<=`, `>=`.

```python
titanic[titanic.age == "18"]
```

```python
titanic[titanic.pclass != 1].pclass.value_counts().plot(kind='bar', color='lightpink')
```

```python
houses[houses['price'] > 5000000]
```

```python
houses[houses['bedrooms'] >= 10 ]
```

```python
houses[houses["sqft_living"] < 1000 ]
```

## The `between` method

* Return boolean series equivalent to `left` `<=` `series` `<=` `right`.
* This function returns a boolean vector containing `True` whenever the corresponding Series element is between the boundary values `left` and `right`.
* `NA` values are treated as `False`.

```python
houses[houses["bedrooms"].between(5,7)]
```

```python
houses[houses["bedrooms"].between(5,7)].bedrooms.value_counts()
```

```python
houses[houses['grade'].between(11,13)]
```

```python
houses[houses['grade'].between(11,13)].grade.value_counts()
```

## The `isin` method

* The `isin()` method checks whether each elements in a Series or DataFrame is contained in a list-like object (list, set, tuple, another Series).
* It returns a `boolean mask` (**`True`**/ **`False`**) that you can use to filter rows.

```python
countries = ["India", "Japan", "South Korea"]
netflix[netflix["country"].isin(countries)]
```

```python
netflix[netflix["country"].isin(countries)].country.value_counts()
```

```python
mature = netflix["rating"].isin(["TV-MA", "R"])
netflix[mature]
```

## Combining conditions using `AND(&)`

Each condition must be wrappped in paranthesis, and the result is a **boolean mask** where only rows that satisfy **all conditions simultaneously** are selected.

```python
women = titanic.sex == 'female'
died = titanic.survived == 0
titanic[women & died]
```

```python
houses[(houses["waterfront"] == 1) & (houses["price"] < 500000)]
```

```python
houses[(houses["view"] == 4) & (houses["grade"] >= 11)]
```

```python
high_quality = houses["grade"] >= 11
good_view = houses["view"] == 4
smaller = houses[high_quality & good_view]
smaller
```

## Combining conditions using `OR (|)`

Combining conditions with OR means the result is **`True`** if at least one condition is true, and **`False`** only if all conditions are false.

```python
houses[(houses["yr_built"] >= 2014) | (houses["yr_renovated"] >= 2014)]
```

```python
netflix[(netflix["director"] == "David Fincher") | (netflix["director"] == "Martin Scorsese")]
```

```python
fincher = netflix["director"] == "David Fincher"
scorsese = netflix["director"] == "Martin Scorsese"
recent = netflix["release_year"] > 2010
netflix[(fincher|scorsese) & recent]
```

## Bitwise Negation

Bitwise negation is a **bit-level operation** that flips every bit in a binary number:
* Each `0` becomes `1`
* Each `1` becomes `0`
  
It is also called the **bitwise NOT** operation

```python
newly_built = houses["yr_built"] >= 2014
newly_renovated = houses["yr_renovated"] >= 2014
recent_homes = newly_built | newly_renovated
houses[~recent_homes]
```

```python
titanic[~(titanic.survived == 0)]
```

```python
titanic[titanic.survived == 0]
```

## `isna` and `notna` methods

* `isna()` - detects missing values (`NaN`) in a DataFrame or Series.It returns boolean mask (`True` where values are missing, `False` otherwise).
* `notna` - opposite of `isna()`. Detects non-missing values.It returns a boolean mask (`True` where values are present, `False` where missing).

```python
netflix[netflix["director"].isna() & netflix["cast"].isna()]
```

```python
netflix[netflix.director.notna()]
```
