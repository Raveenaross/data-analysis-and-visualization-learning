## `sort_values` method 
* It's a **pandas DataFrame method** used to sort rows by the values of one or more columns.
* You can sort in **ascending** (default) or **descending** order.

```python
happiness_report.sort_values(by="Healthy life expectancy")
```

```python
happiness_report.sort_values(by="Healthy life expectancy", ascending=False, inplace=True)
```

```python
happiness_report
```

## Sorting by multiple columns

```python
houses = pd.read_csv("./materials/data/kc_house_data.csv")
titanic = pd.read_csv("./materials/data/titanic.csv")
```

```python
houses.sort_values("price", ascending=False)
```

```python
houses.sort_values(["bedrooms", "bathrooms"], ascending=False)
```

## Sorting text columns

```python
titanic.sort_values("name")
```

In the above the sorting happens in ABC...Zabc..z. So have to change in lower case

`key=` is an argument in `sort_values()`.It let us transform the values in a column before sorting, without actually changing the data.
`lambda col: col.str.lower()` is an **anonymous function** that takes the column and converts all strings to lowercase.

```python
titanic.sort_values("name", key=lambda col: col.str.lower())
```

## `sort_index` method

`.sort_index()` -> Sorts the DataFrame by its **index labels**, not by column values.

```python
happiness_report.sort_index(ascending=False, inplace=True)
```

```python
happiness_report
```

```python
titanic.sort_index(ascending=False)
```

## `loc` function 

`.loc[]` is primarly **label based**, but may also be used with a boolean array.

Allowed inputs are:
* A single label e.g. `5` or `a`
* A list or array of labels e.g. `['a', 'b', 'c']`
* A slice object with labels e.g. `a`:`f`

```python
happiness_report.head()
```

### Accessing a row using labels

```python
happiness_report.loc["Yemen"]
```

```python
happiness_report.loc[["Yemen"]]
```

### Accessing rows using list

```python
happiness_report.loc[["Ireland", "Costa Rica", "United States"]]
```

```python
houses.loc[[7, 50, 800]]
```

### Accessing rows using slice method

```python
titanic.loc[10:16:2]
```

```python
happiness_report.sort_index(inplace=True)
```

```python
happiness_report
```

```python
happiness_report.loc["Denmark":"France"]
```

```python
happiness_report.loc["Denmark":"France":4]
```

## `iloc` function

`iloc[]` is primarly integer position based (from `0` to `length-1` of the axis), but may also be used with a boolean array.

Allowed inputs are:
* An integer, e.g. `5`
* A list or array of integers, e.g. `[4, 3, 0]`
* A slice object with ints , e.g. `1:7`
* A boolean array
* A `callable` function with one argument
* A tuple of row and column indexes. The tuple elements consist of one of the above inputs e.g. `(0, 1)`.

### Accessing a row with a single integer 

```python
happiness_report.iloc[[0]]
```

### Accessing rows with list of integers

```python
happiness_report.iloc[[20,40,100,111]]
```

### Accessing rows using slice operator

```python
happiness_report.iloc[50:60]
```

```python
houses.sort_index(ascending=False, inplace=True)
```

```python
houses
```

```python
houses.iloc[0:5]
```

```python
houses.loc[21612:21611]
```
