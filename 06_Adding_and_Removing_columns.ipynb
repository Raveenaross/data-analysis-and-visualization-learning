# Adding and Removing Columns in a DataFrame

## Dropping Columns

In pandas, the term `drop` refers to a method used to remove rows or columns from a DataFrame or series. Remove rows or columns by specifying label names and corresponding axis, or by directly specifying index or column names.

### Parameters

* `labels`: Single label or list-like of labels to drop.
* `axis`: 0 (rows/index) or 1 (columns). Default is 0.
* `index`: Alternative to labels with axis=0.
* `columns`: Alternative to labels with axis=1.
* `level`: For MultiIndex, specifies which level to drop labels from.
* `inplace`: If True, modifies the DataFrame in place (returns None).
* `errors`: 'raise' (default) throws error if label not found; 'ignore' skips missing labels.

## Adding Static columns

### `insert` method

Insert column into DataFrame at specified location.

#### Parameters
* `lic` - Insertion index.Must verify 0 <= loc <= len(columns)
* `column` - label of the inserted column
* `value` - It can be scalar, series or array-like.Content of the inserted column.
* `allow_duplicates` - bool.Allow duplicate column labels to be created.



```python
titanic["species"] = "human"
titanic
```

```python
houses.insert(0, "county", "King County")
```

```python
houses
```

```python
houses["sale_price"] = houses["price"]
```

```python
houses.head(5)
```

```python
houses.insert(3, "num_bedrooms", houses["bedrooms"])
```

## Creating New "Dynamic" Columns

```python
titanic.drop(columns=["species"], inplace=True)
```

```python
titanic
```

```python
titanic["num_relatives"] = titanic["sibsp"] + titanic["parch"]
```

```python
titanic.head(5)
```

```python
titanic.sort_values("num_relatives", ascending=False)
```

```python
titanic[titanic["survived"] == 1].sort_values("num_relatives", ascending=False)
```

```python
solo_passengers = titanic["num_relatives"] == 0
titanic[solo_passengers].pclass.value_counts()
```

## Finding Largest Bitcoin Price Changes
