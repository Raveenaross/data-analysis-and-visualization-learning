### `unique` & `nunique`

* The `unique()` returns an array of **distinct values** in a Series.
* The `nunique()` returns the **count** of unique values.

```python
houses["bedrooms"].unique()
```

```python
netflix.rating.unique()
```

```python
netflix.rating.nunique()
```

```python
houses.zipcode.unique()
```

```python
houses.zipcode.nunique()
```

```python
netflix.rating.nunique(dropna=False)
```

```python
netflix.rating.nunique(dropna=True)
```

### `nlargest` & `nsmallest`

* `nlargest()` returns the **n largest values** from a Series or DataFrame column.When you want to find the top performers, highest scores, biggest numbers etc.
* `nsmallest()` returns the **n smallest values** from a Series or DataFrame column.When you are looking for the lowest values, smallest cost, earliest dates etc.

```python
houses.price.nlargest(3)
```

```python
houses.price.nsmallest(9)
```

When multiple rows have the same value at the boundary, `keep` decides which of those tied rows are included. 
* `first` - keeps the **first** occurences of the tied values (default behavior).
* `last` - keeps the **last** occurences of the tied values.
* `all` - keeps **all** occurences of the tied values, even if it exceeds `n`.

```python
titanic.pclass.nlargest(5, keep="all")
```

```python
houses.nlargest(10,["price"])
```

```python
houses.nlargest(10, ["bedrooms","bathrooms"])
```

```python
houses.nsmallest(10, ["sqft_lot"])
```

### `value_counts` method
* It is a powerful tool for summarizing the frequency of unique values in a Series.
* **Counts** how many times each unique values appears in a Series.
* **Returns** a Series sorted by count in descending order by default.

```python
houses["bedrooms"].value_counts()
```

```python
houses["bedrooms"].nunique()
```

```python
titanic.sex.value_counts()
```

```python
netflix.director.value_counts().head(10)
```

```python
netflix.director.value_counts(ascending=True)
```

```python
houses["floors"].value_counts()
```

```python
houses[["bedrooms", "bathrooms"]].value_counts()
```

### `plot` method

The `plot()` function in pandas is a convential wrapper around Matplotlib that lets to visualize data from Series or DataFrames.

We can specify the plot type using the `kind` parameter:
* `line` - line plot (default)
* `bar` - vertical bar chart
* `barh` - horizontal bar chart
* `hist` - histogram
* `box` - Box plot
* `kde` - kernel density plot
* `area` - Area plot
* `pie` - Pie chart (Series only)

```python
houses.bedrooms.plot(c="orange")
```

```python
houses.bedrooms.value_counts().plot(kind="bar", color="blueviolet")
```

```python
houses.bedrooms.value_counts().plot(kind="pie")
```

```python
titanic.survived.value_counts().plot(kind="pie")
```

```python
titanic.sex.value_counts().plot(kind="pie")
```

```python
houses[["bedrooms", "bathrooms"]].plot()
```

```python
df = houses[["bedrooms", "bathrooms"]]
df.plot(kind="scatter", x="bedrooms", y="bathrooms", color="hotpink")
```

```python
netflix.rating.value_counts().head(10).plot(kind="bar", color="orangered")
```
